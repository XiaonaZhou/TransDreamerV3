# Boxing

```python
import json
import matplotlib.pyplot as plt
import re

# Function to calculate the average score from a JSONL file
def calculate_average_score(file_path):
    scores = []
    with open(file_path, 'r') as file:
        for line in file:
            data = json.loads(line)
            scores.append(data['episode/score'])
    return sum(scores) / len(scores)

# List of file paths for your three JSONL files
file_paths = ['dreamerv3_atari_boxing.jsonl', 'TDV3_atari_boxing.jsonl']

# Calculate average scores for each file
average_scores = [calculate_average_score(file_path) for file_path in file_paths]


# results from TransDreamerV2 are saved in .txt file
# use another function to get average scores
def calculate_average_score_from_txt(file_path):
    scores = []
    with open(file_path, 'r') as file:
        for line in file:
            match = re.search(r'return (\d+\.\d+)', line)
            if match:
                scores.append(float(match.group(1)))
    return sum(scores) / len(scores)
TDV2_file = 'TDV2_atari_boxing.txt'
average_scores.append(calculate_average_score_from_txt(TDV2_file))

# Bar plot
models = ['DreamerV3', 'TransDreamerV3', 'TransDreamerV2']
plt.bar(models, average_scores, color=['blue', 'green', 'orange'])
plt.ylabel('Average Score')
plt.title('Atari-Boxing after 1.5M steps')
# Show the model names on the bars
for i, model in enumerate(models):
    plt.text(i, 1, model, ha='center', va='bottom', rotation=90, fontsize=20)

plt.ylim(0, max(average_scores) + 2)  # Adjust the y-axis limit for better visualization
# Hide x-axis labels
plt.xticks([])
plt.savefig('plots/Atari-Boxing after 1.5M steps.png')
plt.show()
```

# Freeway
The freeway task takes many step to get positive reward. Took DreamerV3 12M steps to get the first non-zero reward of 3. Waiting on TDV3 to reach 12M steps. TransDreamerV2 is able to get positive reward very soon. 

```python
import json
import matplotlib.pyplot as plt
import re

# Function to calculate the average score from a JSONL file
def calculate_average_score(file_path):
    scores = []
    with open(file_path, 'r') as file:
        for line in file:
            data = json.loads(line)
            scores.append(data['episode/score'])
    return sum(scores) / len(scores)

# List of file paths for your three JSONL files
file_paths = ['dreamerv3_atari_freeway.jsonl', 'TDV3_atari_freeway.jsonl']

# Calculate average scores for each file
average_scores = [calculate_average_score(file_path) for file_path in file_paths]

# results from TransDreamerV2 are saved in .txt file
# use another function to get average scores
def calculate_average_score_from_txt(file_path):
    scores = []
    with open(file_path, 'r') as file:
        for line in file:
            match = re.search(r'return (\d+\.\d+)', line)
            if match:
                scores.append(float(match.group(1)))
    return sum(scores) / len(scores)
TDV2_file = 'TDV2_atari_freeway_with_1M.txt'
average_scores.append(calculate_average_score_from_txt(TDV2_file))

# Bar plot
models = ['DreamerV3', 'TransDreamerV3', 'TransDreamerV2']
plt.bar(models, average_scores, color=['blue', 'green', 'orange'])
plt.ylabel('Average Score')
plt.title('Atari-Freeway')
# Show the model names on the bars
for i, model in enumerate(models):
    plt.text(i, 1, model, ha='center', va='bottom', rotation=90, fontsize=20)

notes = ['after 12M steps', 'after 3M steps', 'after 900k steps']
for i, note in enumerate(notes):
    plt.text(i,average_scores[i] ,note , ha='center', va='bottom')

plt.ylim(0, max(average_scores) + 2)  # Adjust the y-axis limit for better visualization
# Hide x-axis labels
plt.xticks([])
plt.savefig('plots/Atari-Freeway.png')
plt.show()
```
